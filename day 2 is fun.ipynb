```python
# monte carlo simulation to estimate pi
n=100 # number of "darts" to throw
x_list = np.random.uniform(0,1,n) # get a list of random n numbers for x
print("x = ",x_list) 
y_list = np.random.uniform(0,1,n) # get a list of random n numbers for y
print("y = ",y_list)

r_list_squared = (x_list**2) + (y_list**2) # calculate the radius squared
print('r squared = ',r_list_squared) 

r_list = r_list_squared**(1/2) # calculate the radius
print('r = ', r_list)

in_circle = r_list[r_list<=1] # filter the list to only include those in the circle
print(in_circle)

my_pi = 4*len(in_circle)/n # calculate the estimation of pi, len() tells the number of elements in a list
print(my_pi)
```

```python
# numerically solving an equation, iterative method.
def f(r):
    return 10 * 10 ** (2-0.0004*r)
```

```python
r_initial = 1000

r_second = f(r_initial)
print(r_second)

r_third = f(r_second)
print(r_third)

r_fourth = f(r_third)
print(r_fourth)
```

```python
r = 1000
# create a for loop to loop 40 times
for i in range(40):
    r_next = f(r)
    print(r_next)
    r = r_next
```

```python
# fibonacci sequence using list, simpler, more straightforward method.
f_s = [1,1]

for i in range(100):
    last_element = f_s[-1]
    second_last_element = f_s[-2]
    next_sequence = last_element + second_last_element
    f_s.append(next_sequence) # add the new element to the list

print(f_s)
# now, make a loop to repeat the task for 100 times using for loop
```

```python
# get me a function to check if a number is prime
def is_prime(n):
    if n == 1:
        return False
    elif n == 2:
        return True
    else:
        for i in range(2,n):
            if n%i == 0:
                return False
        return True
```
